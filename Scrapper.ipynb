```python
import pandas as pd
import requests #to send http request to a url exp(get,post)
import bs4  #(Beatifulsoup) - to pull data from html and xml files
```

```python
# Getting data Requests

request1 = requests.get('https://www.flipkart.com/poco-c31-royal-blue-32-gb/p/itm9656ebd21e0b3?pid=MOBG73E7ZBHHQ3RB&lid=LSTMOBG73E7ZBHHQ3RBQ7POR7&marketplace=FLIPKART&q=poco+c31&store=tyy%2F4io&srno=s_1_1&otracker=AS_Query_OrganicAutoSuggest_3_6_na_na_ps&otracker1=AS_Query_OrganicAutoSuggest_3_6_na_na_ps&fm=Search&iid=928ef87e-9040-41b0-91de-273876d4f120.MOBG73E7ZBHHQ3RB.SEARCH&ppt=sp&ppn=sp&ssid=qpn9znbppc0000001685245664671&qH=8a1462a8a580f1cb')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
# Fetching Reviews and Comments

reviews = soup.find_all('div',{'class':'t-ZTKy'});
for review in reviews:
    print(review.get_text()+ '\n\n')
```

```python
# Fetching Average Rating

ratings = soup.find('div',{'class':'_3LWZlK'})
for rate in ratings:
    print(rate.get_text())
```

```python
# Overall Rating

rating = soup.find_all('div',{'class':'_3LWZlK _1BLPMq'})
for rate in rating:
    print(rate.get_text(),'\n')
```

```python
#Tags

tags = soup.find('span',{'class':'yhB1nd GXgmTe'})
print(tags.get_text())
```

```python
#Customer Name

names = soup.find_all('p',{'class':'_2sc7ZR _2V5EHH'})
for name in names:
    print(name.get_text(),'\n')
```

```python
# Extra
pricing = soup.find('div',{'class':'_30jeq3 _16Jk6d'}).get_text()
pricing
```

```python
df2 = requests.get("https://www.flipkart.com/search?q=mobile+under+12000+rs&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_5_15_na_na_ps&otracker1=AS_Query_OrganicAutoSuggest_5_15_na_na_ps&as-pos=5&as-type=HISTORY&suggestionId=mobile+under+12000+rs&requestId=2c72c16e-e169-4b06-b968-c659eb405ddd")
df2
```

```python
soup1 = bs4.BeautifulSoup(df2.text)
soup1
```

```python
name_data = soup1.find_all('div',{'class':'_4rR01T'})
price_data = soup1.find_all('div', {'class' : '_30jeq3 _1_WHN1'})
phn_name=[]
phn_price=[]
```

```python
for item in name_data:
    phn_name.append(item.text)

phn_name
```

```python
for item in price_data:
    phn_price.append(item.text)

phn_price
```

```python
df = {'phone Name' : phn_name,
      'phone price' : phn_price}
final_df = pd.DataFrame(df)
final_df
```
